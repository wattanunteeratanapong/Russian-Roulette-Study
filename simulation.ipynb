```python
# Russian Roulette (Sequence firing 1v1)
import random
import matplotlib.pyplot as plt
```

```python
# Simulation
times = 10000000
first_death = 0
second_death = 0
stats = []

for i in range(times):
    chamber = [1, 0, 0, 0, 0, 0]

    random.shuffle(chamber)

    i = 1
    while True:
        bullet = chamber.pop(0)
        if bullet == 1:
            first_death += 1
            stats.append(i)
            break
        i += 1

        bullet = chamber.pop(0)
        if bullet == 1:
            second_death += 1
            stats.append(i)
            break
        i += 1
```

```python
print(f"First player died {first_death} times")
print(f"Second player died {second_death} times")
print(f"Average turns to die: {sum(stats) / len(stats)}")
```

```python
# Plotting the statistics
total_deaths = first_death + second_death
probabilities = [first_death / total_deaths, second_death / total_deaths]

labels = ['First Player', 'Second Player']
plt.bar(labels, probabilities, color=['blue', 'red'])
plt.title('Probability of Death for Each Player in Russian Roulette Simulation')
plt.xlabel('Player')
plt.ylabel('Probability')
plt.show()
```

```python
# Plotting the statistics
plt.hist(stats, bins=range(1, max(stats) + 2), edgecolor='black', width=0.8, density=True)
plt.title('Distribution of Turns to Death in Russian Roulette Simulation')
plt.xlabel('Number of Turns')
plt.ylabel('Probability')
plt.show()
```
